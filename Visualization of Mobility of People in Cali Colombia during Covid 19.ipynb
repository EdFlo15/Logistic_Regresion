```python
import pandas as pd
import matplotlib.pyplot as plt # Esta función es utilizada para mostrar los gráficos
```

```python
df=pd.read_csv("Global_Mobility_Report_actualizado_20220303.csv")
```

```python
df.head()
```

```python
df2=df[df['sub_region_2']=='Cali']
```

```python
df2['sub_region_2'].value_counts()
```

```python
df2.shape
```

```python
df2.tail()
```

```python
df2.set_index('date', inplace=True)
```

```python
df2.index=pd.to_datetime(df2.index)
```

```python
df2.shape
```

```python
columnas=['retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline','parks_percent_change_from_baseline',
'transit_stations_percent_change_from_baseline','workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']
diario=df2[columnas].resample('1D').mean()
#semanal=df3[columnas].resample('W').mean()
```

```python
inicio, final='2020-02-15','2022-03-02'
fig, eje=plt.subplots(figsize=(20,10))
eje.plot(diario.loc[inicio:final, 'retail_and_recreation_percent_change_from_baseline'],marker='.',label='Comercio minorista y recreación')
eje.plot(diario.loc[inicio:final, 'grocery_and_pharmacy_percent_change_from_baseline'],marker='.',label='Comestibles y farmacias')
eje.plot(diario.loc[inicio:final, 'parks_percent_change_from_baseline'],marker='.',label='Parques')
eje.plot(diario.loc[inicio:final, 'transit_stations_percent_change_from_baseline'],marker='.',label='Estaciones de transporte')
eje.plot(diario.loc[inicio:final, 'workplaces_percent_change_from_baseline'],marker='.',label='lugares de trabajo')
eje.plot(diario.loc[inicio:final, 'residential_percent_change_from_baseline'],marker='.',label='Cambio de residencia')
eje.set_ylabel("Cambio en porcentaje desde la base de referencia",size=14)
eje.set_title("¿Como ha cambiado la movilidad de diferentes sectores economicos en Cali a causa del COVID-19?", size=18)
eje.set_xlabel("Linea de tiempo",size=14)
#eje.axvline(pd.to_datetime('2020-03-25'), linestyle ='--', color='#A4A4A4', linewidth=2)
eje.axhline(0, linestyle ='--', color='#A4A4A4', linewidth=2)

#eje.set_facecolor('#eafff5')

""" eje.annotate("Se declara Covid 19 \n como una pandemia",
                  xy=(pd.to_datetime('2020-03-25'), 28),
                  xytext=(pd.to_datetime('2020-05-10'), 25),
                  size=13, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="-|>",
                                  connectionstyle="arc3,rad=-0.2",
                                  fc="w"),
                  ) """

""" eje.annotate("Segundo pico \n de la pandemia",
                  xy=(pd.to_datetime('2021-01'), -90),
                  xytext=(pd.to_datetime('2021-01'), -90),
                  size=13, va="center", ha="center")

eje.annotate("Tercer pico \n de la pandemia",
                  xy=(pd.to_datetime('2021-05'), -82),
                  xytext=(pd.to_datetime('2021-05'), -82),
                  size=13, va="center", ha="center") """

eje.legend()
```
