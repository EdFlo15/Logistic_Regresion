```python
from abc import abstractclassmethod
from ftplib import FTP
import ftplib
import os
import os.path
import shutil
from os import getcwd, remove
import tkinter
from tkinter import StringVar
from tkinter import StringVar
```

```python
### Elaboramos la aplicación:
ventana=tkinter.Tk()
ventana.geometry("900x300")


etiqueta=tkinter.Label(ventana,text="Botón confirmación", pady=10, font="Verdana 15")
etiqueta.grid(row=0,column=0)
# poner en pantalla un rtexto de salida de la aplicación:
#label_confirmacion1=tkinter.Label(ventana, text=imprimir_pantalla())

# para cada rem se debe crear un Label con el fin de mostrar el mensaje de exito o falla 
label_confirmacion1=tkinter.Label(ventana, text="",fg="green")
label_confirmacion1.grid(row=2, column=0)

label_confirmacion2=tkinter.Label(ventana, text="",fg="green")
label_confirmacion2.grid(row=3, column=0)

label_confirmacion3=tkinter.Label(ventana, text="",fg="green")
label_confirmacion3.grid(row=4, column=0)

# definimos la funcion generica para mostrar el mensaje exitoso de transferencia.
def msg_exitoso(mensaje, label):
    texto=StringVar()
    texto.set(mensaje)
    label.config(textvariable=texto, fg="green")

# definimos la funcion generica para mostrar el mensaje de error capturado por una excepción.

def msg_fallido(error, remmiter, label):
    texto=StringVar()
    texto.set(remmiter+" : "+str(error))
    label.config(textvariable=texto, fg="red")


# definimos una función de carga de archivos hacia el FTP, recibe los datos y la ruta en el FTP para relizar la carga
def cargar_archivos(lista,ruta,ftp):
    for n in range(0,len(lista)):
     ftp.storbinary('STOR ' + lista[n], open(ruta+"/"+lista[n], 'rb'))

# funcion para sacar el backup de los archivos. Define un destino en donde se alojan los archivos de respaldo
# importante posicionarse en el origen.
def realizar_backup(files, ruta_origen, remmiter):
    destino="/Users/****/Desktop/backup/"+remmiter
    #print(len([name for name in os.listdir(destino) if os.path.isfile(os.path.join(destino, name))]))
    #print(destino)
    os.chdir(ruta_origen)
    for n in files:
        shutil.copy2(n,destino)

# me retorna el numero de archivos en la carpeta. Importante definri el numero de archivos a mantener, un buen numero puede ser 1.000: Ejemplo y pruebas 5
def depurar_backup(remmiter):
    destino="/Users/*****/Desktop/backup/"+remmiter
    lista_files=os.listdir(destino)
    os.chdir(destino)
    if len(lista_files)>5:
        num_max_borrar=len(lista_files)-5
        for m in range(0, num_max_borrar):
            #print(lista_files[m])
            remove(lista_files[m])


# definimos una funcion que borra los archivos de la carpeta origen
# importante el uso de la función os.chdir, para establecer la ruta de eliminación.

def eliminar_archivos(ficheros, ruta_eliminar):
    os.chdir(ruta_eliminar)
    for n in ficheros:
        remove(n)

# esta es la funcion main, creada para correr todo el proceso a través de la aplicación.
def ejecutar():
    # establecemos una ruta raiz desde donde vamoa a trabajar.
    folder="/Users/****/*****/proyectos visual studio/conexion_FTP/archivos"
    #recorremos los directorios
    for ruta, carpetas, archivos in os.walk(folder, topdown=True):
        for carp in carpetas:
            ruta_temp=folder+"/"+carp
            print(ruta_temp)
            with os.scandir(ruta_temp) as ficheros:
                ficheros=[fichero.name for fichero in ficheros if fichero.is_file()]
                print(ficheros)
                if carp=="356":
                    remmiter="****"
                    try:
                       ftp=FTP(host='******')
                       print('respuesta desde el servidor' + ftp.getwelcome())
                       ftp.login(user='****',passwd='****')
                       print("conexion exitosa")
                       ftp.cwd('OUT')
                       cargar_archivos(ficheros,ruta_temp,ftp)
                       ftp.quit
                       depurar_backup(remmiter)
                       realizar_backup(ficheros, ruta_temp, remmiter)
                       eliminar_archivos(ficheros,ruta_temp)
                       mensaje=remmiter + ": transferencia exitosa al servidor FTP"
                       msg_exitoso(mensaje,label_confirmacion1)
                    
                    except ftplib.all_errors as e:
                          print( 'Error -> ', e )
                          print(type(e))
                          msg_fallido(e, remmiter,label_confirmacion1)
        
                elif carp=="456":
                    remmiter="********"
                    try:
                       ftp=FTP(host='127.0.0.1')
                       print('respuesta desde el servidor' + ftp.getwelcome())
                       ftp.login(user='******',passwd='*******')
                       print("conexion exitosa")
                       ftp.cwd('OUT')
                       cargar_archivos(ficheros,ruta_temp,ftp)
                       ftp.quit
                       depurar_backup(remmiter)
                       realizar_backup(ficheros, ruta_temp, remmiter)
                       eliminar_archivos(ficheros,ruta_temp)
                       mensaje= remmiter+": transferencia exitosa al servidor FTP"
                       msg_exitoso(mensaje, label_confirmacion2)
                    except ftplib.all_errors as e:
                        print( 'Error -> ', e )
                        msg_fallido(e,remmiter,label_confirmacion2)

                elif carp=="645":
                    remmiter="********"
                    try:
                       ftp=FTP(host='127.0.0.1')
                       print('respuesta desde el servidor' + ftp.getwelcome())
                       ftp.login(user='*********',passwd='***********')
                       print("conexion exitosa")
                       ftp.cwd('OUT')
                       cargar_archivos(ficheros,ruta_temp,ftp)
                       ftp.quit
                       depurar_backup(remmiter)
                       realizar_backup(ficheros, ruta_temp, remmiter)
                       eliminar_archivos(ficheros,ruta_temp)
                       mensaje= remmiter+": transferencia exitosa al servidor FTP"
                       msg_exitoso(mensaje, label_confirmacion3)
                    except ftplib.all_errors as e:
                        print( 'Error -> ', e )
                        msg_fallido(e,remmiter,label_confirmacion3)

boton1=tkinter.Button(ventana, text="Confirmaciones \n automaticas", bg="blue", fg="white", bd=4, height=2, width=20, activebackground="yellow",
command= lambda: ejecutar())
boton1.grid(row=1, column=0)


ventana.mainloop()
```
